```python
"""
Linear Regression
1- Find a pattern in the old data
2- Straight line
3- Line
4- y = mx + b

[[value]] is a 2D array, because the model expects a 2D array as input

X = [[1], [2], [3], [4], [5]]
y = [2, 4, 6, 8, 10]

1- It's not about the line, its about the story
2- 5 rows use start
3- accuracy is not the goal

"""


from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4], [5]]
y = [40, 50, 65, 75, 90]

model = LinearRegression()

model.fit(X, y)
hours = float(input("Enter the number of hours: "))

predicted_marks = model.predict([[hours]])

print(f"Predicted marks for {hours} hours: {predicted_marks[0]:.2f}")

```

```python
"""
Logistic Regression - Classification

1- Find a pattern in the old data
2- y = 1 / (1 + e^(-z)) where z = mx + b
3- 0 to 1 yes or no

"""

from sklearn.linear_model import LogisticRegression

X = [[1], [2], [3], [4], [5]] # hours studied
y = [0, 0, 1, 1, 1]  # o for fail, 1 for pass

model = LogisticRegression()

model.fit(X, y)
hours = float(input("Enter the number of hours: "))

result = model.predict([[hours]])[0]

if result == 1:
    print(f"Predicted result for {hours} hours: Pass")
else:
    print(f"Predicted result for {hours} hours: Fail")

```

```python
"""
KNN: K-Nearest Neighbors - Classification

1- Find a pattern in the old data
2- Look at the nearest neighbors and take a vote
3- K is the number of neighbors to look at
4- 0 to 1 yes or no

X= [
    [180, 7], # weight, size
    [160, 6], 
    [170, 6.5],
    [150, 5.5],
    [190, 7.5]
]

y = [0, 0, 0, 1, 1] # 0 for apple, 1 for orange

formula: distance = sqrt((x2 - x1)^2 + (y2 - y1)^2)

"""

from sklearn.neighbors import KNeighborsClassifier

X= [
    [180, 7],
    [200, 7.5],
    [250, 8],
    [300, 8.5],
    [330, 9],
    [360, 9.5]
]

y = [0, 0, 0, 1, 1, 1] # 0 for apple, 1 for orange

model = KNeighborsClassifier(n_neighbors=3)

model.fit(X, y)

weight = float(input("Enter the weight in grams: "))
size = float(input("Enter the size in cm: "))

result = model.predict([[weight, size]])[0]

if result == 0:
    print(f"Predicted result for weight {weight} and size {size}: Apple")
else:
    print(f"Predicted result for weight {weight} and size {size}: Orange")


 
```

```python
"""
Decision Tree - Classification

1- Find a pattern in the old data
2- Create a tree based on the features
3- 0 to 1 yes or no
4- If the feature is less than a certain value, go left, otherwise go right
5- The leaf nodes are the predicted class
6- The internal nodes are the features
7- The root node is the feature that best splits the data
8- The best split is the one that minimizes the Gini impurity or maximizes the information gain

is fever > 100 
    yes > cough?
        yes > flu
        no > maybe
    no - no Flu

"""

from sklearn.tree import DecisionTreeClassifier

X = [
    [7, 2],
    [8, 3],
    [9, 8],
    [10, 9]
]

y= [0, 0, 1, 1] # 0 for apple, 1 for orange

model = DecisionTreeClassifier()

model.fit(X, y)

size = float(input("Enter the fruit size in cm: "))
shade = float(input("Enter the color shade (1-10): "))

result = model.predict([[shade, size]])[0]

if result == 0:
    print(f"Predicted result for size {size} and shade {shade}: Apple")
else:
    print(f"Predicted result for size {size} and shade {shade}: Orange")
```
