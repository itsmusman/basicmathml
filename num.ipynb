1: Basic Probability (Events and Outcomes)

Outcome: A single Result 
Example: Dice
1,2,3,4,5,6 --> Each is an outcome

Event: A group of Outcomes
Example: Getting and even number --> {2,4,6}

Dice Roll:
Event: Get even number

favorble outcomes = 2,4,6 --> 3 outcomes
Total outcomes = 6

Probability = 3/6 = 1/2 = 0.5

------------------------------------------------------

2: Conditional Probability (Dependent Events)

This is when one event affects another

what is the chance of B, given than A already happened?

Example: You have three balls 2 Red 1 Blue:

Pick one ball don't put it back

Step1: First pick is red

probability = 2/3

Step2: now only 2 balls are left:
1 Red and 1 bLue

Probability of Red again = 1/2

This changed --> so event are dependent

Formula:

P(A|B) = P(A and B)/P(B)

P = Probability
A = The event we want 
| = Given that
B = ALready Happened

--------------------------------------------------------

3: Random Variables:

A random variable is a number that represent the outcomes of a random event 

Real world   -->       Random Variable

2 Types of Random Variables:

Discrete: Countable values, can you count it?  no of students, no of call

continuous: Any value in Range, Infinite Possibilities? can you measue it? height, weight, temprature

```python
import random
count_head= 0

for i in range(100):
    if random.choice(["Head", "Tail"]) == "Head":
        count_head += 1
print("Heads", count_head)        
```

```python
balls = ["Red", "Red", "Red", "Blue", "Blue"]

#pick 1st ball

first = random.choice(balls)
balls.remove(first)

#pick 2nd ball

second = random.choice(balls)

print("First", first)
print("Second", second)
```

```python
#without simulatoin
result = random.choice(["Head", "Tail"])
print(result)
```

```python
#Simulation
head = 0
tail = 0

for i in range(100):
    result = random.choice(["Head", "Tail"])
    if result == "Head":
        head += 1
    else:
        tail += 1
print("Heads", head)
print("Tails", tail)
```

```python
Simulation
head = 0
total = 1000

for i in range(total):
    if random.choice(["Head", "Tail"]) == "Head":
        head +=1

probability = head/total

print("Estimated Probability of Head", probability)
```
