```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

data = pd.read_csv('student.csv')

X = data[['Hours']]  # double bracket = 2d input
y = data['Score']   # target column

model = LinearRegression()
model.fit(X, y)

predicted_scores = model.predict(X)

#evaluate the model

mae= mean_absolute_error(y, predicted_scores)
mse= mean_squared_error(y, predicted_scores)
rmse = np.sqrt(mean_squared_error(y, predicted_scores))

#show the results
print(f'Mean Absolute Error: {mae:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')


#predicting the score for 9.25 hours of study
new_hours = 9.25
new_predicted_score = model.predict(np.array([[9.25]]))

print(f'Predicted score for {new_hours} hours of study: {new_predicted_score[0]:.2f}')

```
